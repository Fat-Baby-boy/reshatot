```python
import socket
import threading
import time

# הגדרות חיבור
HOST = '127.0.0.1'
PORT = 12345

def run_server():
    """שרת פשוט שמקבל הודעות ומחזיר אישור"""
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        s.bind((HOST, PORT))
        s.listen()
        print(f"[Server] מאזין ב-{HOST}:{PORT}...")
        
        conn, addr = s.accept()
        with conn:
            while True:
                data = conn.recv(1024)
                if not data:
                    break
                print(f"[Server] התקבלה הודעה: {data.decode()}")
                conn.sendall(b"Message Received")

def run_client():
    """לקוח ששולח הודעות שלום פשוטות"""
    time.sleep(1)  # המתנה קלה לעליית השרת
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        try:
            s.connect((HOST, PORT))
            
            for i in range(1, 5):
                msg = "Hello World"+" "+str(i)
                print(f"[Client] שולח: {msg}")
                s.sendall(msg.encode())
                
                resp = s.recv(1024)
                print(f"[Client] תשובה: {resp.decode()}")
                time.sleep(1)
                
        except ConnectionRefusedError:
            print("[Client] שגיאה: השרת לא זמין.")

if __name__ == "__main__":
    # הרצת השרת והלקוח במקביל
    threading.Thread(target=run_server, daemon=True).start()
    run_client()
```

```python
import socket

def start_listener():
    # הגדרות (חייבות להתאים ל-dst_port שבקוד השולח שלך)
    host = '127.0.0.1'
    port = 12345
    
    # יצירת הסוקט
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        # מאפשר הרצה חוזרת של הקוד בלי לחכות שהפורט יתנקה מהמערכת
        s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        
        s.bind((host, port))
        s.listen(1)
        print(f"Listener מאזין ב-{host}:{port}...")
        
        while True:
            conn, addr = s.accept()
            with conn:
                print(f"התחברות חדשה מ: {addr}")
                while True:
                    data = conn.recv(1024)
                    if not data:
                        break
                    # הדפסת המידע שהתקבל כטקסט
                    print(f"התקבל ב-Listener: {data.decode('utf-8', errors='ignore')}")
                    
                    # שליחת אישור חזרה (ACK ברמת האפליקציה)
                    conn.sendall(b"Got it!")

if __name__ == "__main__":
    start_listener()
```
